## Processing Dataset (Urdu-Poetry)

```python

from spacy import blank

# import urduhack 
from urduhack.normalization import normalize
from urduhack.preprocessing import normalize_whitespace 
from urduhack.preprocessing import remove_accents

# import pandas
import pandas as pd

# import sklearn
from sklearn import svm  
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report, confusion_matrix
import csv
import pandas as pd
import numpy as np
# read dataset from file

filename = "/home/manzur/Documents/Urdu/Urdu_corp_source.csv"
urdu_source = pd.read_csv(filename, encoding="utf8", usecols=['Data','Label'])
urdu_source = urdu_source.dropna() #removing 'NaN' values from data
data = urdu_source['Data']

labels = pd.unique(urdu_source['Label']).tolist()

urdu_dict = {}

for label in labels:
    label_data = pd.unique(urdu_source[urdu_source['Label'] == label]['Data'])
    # convert labeled data to string
    string_data = ' '.join([str(elem) for elem in label_data])

    #normalization
    urdu_source['Data'] = urdu_source['Data'].apply(normalize)
    urdu_source['Data'] = urdu_source['Data'].apply(remove_accents)
    urdu_source['Data'] = urdu_source['Data'].apply(normalize_whitespace)
   

    # create blank document
    word_blank_page = blank('ur')

    # place dataset into document
    doc = word_blank_page(string_data)

    text = doc.text

     #clean text
    text = text.replace('[', '')
    text = text.replace(']', '')
    text = text.replace("'", '')

    # convert to list of words
    words = text.split(' ')
    print("{} words: {}".format(label, len(words)))
    
    #urdu_dict[label] = words
```

```python
with open("/home/manzur/Downloads/stopwords-ur.txt", 'r', encoding='utf8') as f:
    STOP_WORDS = f.read().replace("\n", " ")
def remove_stopwords(text: str):
   return " ".join(word for word in text.split() if word not in STOP_WORDS)
urdu_source['Data'] = urdu_source['Data'].apply(remove_stopwords)
urdu_source['Data']
```

```python
#Splitting dataset into training and testing parts.
X_train, X_test, Y_train, Y_test = train_test_split(urdu_source['Data'], urdu_source['Label'],test_size = 0.20, random_state = 7, shuffle = True)
#applying 
max_feature_num = 5000
vectorizer = TfidfVectorizer(max_features=max_feature_num)
train_vecs = vectorizer.fit_transform(X_train)
test_vecs = TfidfVectorizer(max_features=max_feature_num, vocabulary=vectorizer.vocabulary_).fit_transform(X_test)

#plotting in dataframe
pd.DataFrame(test_vecs.toarray(), columns=vectorizer.get_feature_names())
```

```python
import joblib
from joblib import dump
from sklearn.pipeline import Pipeline
def SVM_classifier(train_vecs, Y_train, test_vecs, Y_test):
    # Training
    SVM = svm.LinearSVC(max_iter=100)
    SVM = SVM.fit(train_vecs, Y_train)

    # Testing
    test_predictionSVM = SVM.predict(test_vecs)
    #pipeline of tfidf and best classifier SVM
    pipeline = Pipeline(steps= [('tfidf',vectorizer),
                            ('model', svm.LinearSVC(max_iter=100))])

# fit the pipeline model with the training data                            
    pipeline.fit(X_train, Y_train)
    dump(pipeline,filename="pipelined.joblib")
    
    return classification_report(test_predictionSVM, Y_test), confusion_matrix(test_predictionSVM, Y_test)

class_report , conf_matrix = SVM_classifier(train_vecs, Y_train, test_vecs, Y_test)
print('Results of SVM CLASSIFIER on TF-IDF Vectorizer\n')
print(class_report)
print(conf_matrix)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

def RandomForest_classifier(train_vecs, Y_train, test_vecs, Y_test):
    # Training
    RF = RandomForestClassifier(max_depth=5, n_estimators=10)
    RF.fit(train_vecs, Y_train)

    # Testing
    test_predictionRF = RF.predict(test_vecs)
    return classification_report(test_predictionRF, Y_test), confusion_matrix(test_predictionRF, Y_test)

class_report , conf_matrix = RandomForest_classifier(train_vecs, Y_train, test_vecs, Y_test)
print('Results of RandomForest CLASSIFIER on TF-IDF Vectorizer\n')
print(class_report)
```

# Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
def NaiveBayes_classifier(train_vecs, Y_train, test_vecs, Y_test):
    # Training
    NB = GaussianNB()
    NB.fit(train_vecs.toarray(), Y_train)

    # Testing
    test_predictionNB = NB.predict(test_vecs.toarray())
    return classification_report(test_predictionNB, Y_test), confusion_matrix(test_predictionNB, Y_test)

class_report , conf_matrix = NaiveBayes_classifier(train_vecs, Y_train, test_vecs, Y_test)
print('Results of NaiveBayes CLASSIFIER on TF-IDF Vectorizer\n')
print(class_report)
```

# Desicion Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
import joblib
from joblib import dump
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
def DecisionTree_classifier(train_vecs, Y_train, test_vecs, Y_test):
    # Training
    clf = DecisionTreeClassifier(max_depth=10)
    clf = clf.fit(train_vecs, Y_train)

    # Testing
    test_prediction = clf.predict(test_vecs)
 
    

    return classification_report(test_prediction, Y_test), confusion_matrix(test_prediction, Y_test)

class_report , conf_matrix = DecisionTree_classifier(train_vecs, Y_train, test_vecs, Y_test)
print('Results of DecisionTree CLASSIFIER on TF-IDF Vectorizer\n')
print(class_report)
```

# Logistic Regression Classifier

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
def LogisticRegression_classifier(train_vecs, Y_train, test_vecs, Y_test):
    # Training
    clf = OneVsRestClassifier(LogisticRegression())
    clf.fit(train_vecs.toarray(), Y_train)

    # Testing
    test_prediction = clf.predict(test_vecs.toarray())
    return classification_report(test_prediction, Y_test), confusion_matrix(test_prediction, Y_test)

class_report , conf_matrix = LogisticRegression_classifier(train_vecs, Y_train, test_vecs, Y_test)
print('Results of LogisticRegression CLASSIFIER on TF-IDF Vectorizer\n')
print(class_report)
```

```python
from urduhack.models.lemmatizer import lemmatizer
def lemitizeStr(str):
    lemme_str = ""
    temp = lemmatizer.lemma_lookup(str)
    for t in temp:
        lemme_str += t[0] + " "
    
    return lemme_str
urdu_source['lemmatized_Data'] = urdu_source['Data'].apply(lemitizeStr)
```
